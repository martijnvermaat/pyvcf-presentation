```python
results = open('results.csv')
headers = next(results)[1:-1].split('\t')
data = [dict(zip(headers, line.rstrip().split('\t'))) for line in results]

pylab.rcParams['ytick.major.pad']='18'

figure(figsize=(20, 9))

bar_width = 0.3

index = arange(len(data) / 3)

plt.barh(index - bar_width,
        [float(d['time'].split()[3]) for d in data if d['callstats'] == 'no'],
        bar_width,
        color='#eeeeee',
        label='no callstats')
plt.barh(index,
        [float(d['time'].split()[3]) for d in data if d['callstats'] == 'some'],
        bar_width,
        color='#eeaaaa',
        label='some callstats')
plt.barh(index + bar_width,
        [float(d['time'].split()[3]) for d in data if d['callstats'] == 'all'],
        bar_width,
        color='#ee6666',
        label='all callstats')

plt.yticks(index + 0.5 * bar_width,
           list(reversed(['CyVCF',
            'PyVCF lazy (Cython)',
            'PyVCF lazy',
            'PyVCF (Cython)',
            'PyVCF'])),
           size=36)
grid(True, axis='x')
plt.xticks(range(0, 400, 50), [])
plt.ylim(-0.35, 4.6)
xlim(0, 600)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().yaxis.set_tick_params(size=0)
plt.gca().xaxis.set_tick_params(size=0)
legend(loc='upper right', fontsize=36)
plt.tight_layout()
plt.savefig('performance.png', format='png')
```
